````python
import gc

import torch.nn as nn
import torch.nn.functional as F
from fastai.callbacks import children, hook_outputs
from fastai.utils.mem import gpu_mem_get_free_no_cache
from fastai.vision import (
    DatasetType,
    ImageImageList,
    LossMetrics,
    NormType,
    get_transforms,
    imagenet_stats,
    models,
    requires_grad,
    unet_learner,
)
from torchvision.models import vgg16_bn
````

````python
# define our general parameters and input paths/data
bs, size = 32, 128
arch = models.resnet34

path_lr = "/content/gdrive/My Drive/toupload/input/"
path_hr = "/content/gdrive/My Drive/toupload/target/"

src = ImageImageList.from_folder(path_lr).split_by_rand_pct(0.1, seed=42)
````

````python
# this function is used multiple time for preparing our data at various sizes
def get_data(bs, size):
    data = (
        src.label_from_func(lambda x: f"{path_hr}/{x.name}")
        .transform(get_transforms(max_zoom=2.0), size=size, tfm_y=True)
        .databunch(bs=bs)
        .normalize(do_y=True)
    )

    data.c = 3
    return data
````

````python
# create and display our initial data ("blurry" sentinel-2 Level 2A and high resolution Digital Globe imagery)
data = get_data(bs, size)
data.show_batch(ds_type=DatasetType.Valid, rows=2, figsize=(5, 5))
````

## Feature loss

This section defines the necessary feature loss function as described in the video and paper linked above.

````python
def gram_matrix(x):
    n, c, h, w = x.size()
    x = x.view(n, c, -1)
    return (x @ x.transpose(1, 2)) / (c * h * w)
````

````python
base_loss = F.l1_loss
````

````python
vgg_m = vgg16_bn(True).features.cuda().eval()
requires_grad(vgg_m, False)
````

````python
blocks = [i - 1 for i, o in enumerate(children(vgg_m)) if isinstance(o, nn.MaxPool2d)]
blocks, [vgg_m[i] for i in blocks]
````

````python
class FeatureLoss(nn.Module):
    def __init__(self, m_feat, layer_ids, layer_wgts):
        super().__init__()
        self.m_feat = m_feat
        self.loss_features = [self.m_feat[i] for i in layer_ids]
        self.hooks = hook_outputs(self.loss_features, detach=False)
        self.wgts = layer_wgts
        self.metric_names = (
            [
                "pixel",
            ]
            + [f"feat_{i}" for i in range(len(layer_ids))]
            + [f"gram_{i}" for i in range(len(layer_ids))]
        )

    def make_features(self, x, clone=False):
        self.m_feat(x)
        return [(o.clone() if clone else o) for o in self.hooks.stored]

    def forward(self, input, target):
        out_feat = self.make_features(target, clone=True)
        in_feat = self.make_features(input)
        self.feat_losses = [base_loss(input, target)]
        self.feat_losses += [base_loss(f_in, f_out) * w for f_in, f_out, w in zip(in_feat, out_feat, self.wgts)]
        self.feat_losses += [
            base_loss(gram_matrix(f_in), gram_matrix(f_out)) * w**2 * 5e3
            for f_in, f_out, w in zip(in_feat, out_feat, self.wgts)
        ]
        self.metrics = dict(zip(self.metric_names, self.feat_losses))
        return sum(self.feat_losses)

    def __del__(self):
        self.hooks.remove()
````

````python
feat_loss = FeatureLoss(vgg_m, blocks[2:5], [5, 15, 2])
````

## Train

Using the aforementioned loss function, we can train a model to predict the high resolution images from the medium resolution Sentinel-2 imagery.

````python
# define our model as a unet with the above loss function
wd = 1e-3
learn = unet_learner(
    data, arch, wd=wd, loss_func=feat_loss, callback_fns=LossMetrics, blur=True, norm_type=NormType.Weight
)
gc.collect();
````

````python
# search the learning rate space
learn.lr_find()
learn.recorder.plot()
````

````python
# define our learning rate at a position of reasonable gradient descent
lr = 1e-3
````

````python
# define a helper function for training 10 epochs, saving the model, and displaying the results
def do_fit(save_name, lrs=slice(lr), pct_start=0.9):
    learn.fit_one_cycle(10, lrs, pct_start=pct_start)
    learn.save(save_name)
    learn.show_results(rows=1, imgsize=5)
````

````python
# train
do_fit("1a", slice(lr * 10))
````

````python
# unfreeze the encoder layers to fine tune our model
learn.unfreeze()
````

````python
do_fit("1b", slice(1e-5, lr))
````

````python
# create new data with higher resolution targets (and a smaller batch to avoid memory errors)
data = get_data(8, size * 2)
````

````python
# add our new data to the fastai learner and re-freeze the encoder layers
learn.data = data
learn.freeze()
gc.collect()
````

````python
learn.load("1b");
````

````python
# train again using the higher resolution targets
do_fit("2a")
````

````python
learn.unfreeze()
````

````python
do_fit("2b", slice(1e-6, 1e-4), pct_start=0.3)
````

## Evaluation and Prediction

Now that we have a trained network, let's see how it works.

````python
# display predictions on the input training data
learn.show_results(rows=10, imgsize=5)
````

````python
# reset the learner
learn = None
gc.collect();
````

````python
# the max size of the test image depends on the available GPU RAM
free = gpu_mem_get_free_no_cache()
if free > 8000:
    size = (1280, 1600)  # >  8GB RAM
else:
    size = (820, 1024)  # <= 8GB RAM
print(f"using size={size}, have {free}MB of GPU RAM free")
````

````python
# load our learner again
learn = unet_learner(data, arch, loss_func=F.l1_loss, blur=True, norm_type=NormType.Weight)
````

````python
# add input data to predict on
data_input = (
    ImageImageList.from_folder(path_lr)
    .split_by_rand_pct(0.1, seed=42)
    .label_from_func(lambda x: path_hr / x.name)
    .transform(get_transforms(), size=size, tfm_y=True)
    .databunch(bs=1)
    .normalize(imagenet_stats, do_y=True)
)
data_input.c = 3
````

````python
# load our model
learn.load("2a")
learn.data = data_input
````

````python
# display final predictions
learn.show_results()
````
